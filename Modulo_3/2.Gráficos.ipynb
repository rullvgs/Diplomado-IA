```python
experiencia = [1, 2, 4, 5, 6, 7, 12]
salario_data_scientis = [2345, 5622, 6578, 8533, 3923, 9999, 4567]
salario_ing_software = [5422, 3235, 6444, 3001, 4311, 5432, 4422]
```

```python
plt.plot(experiencia, salario_data_scientis)
plt.plot(experiencia, salario_ing_software)
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.show()
```

```python
plt.plot(experiencia, salario_data_scientis)
plt.plot(experiencia, salario_ing_software)
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.legend(['Data Scientist', 'Ing Software'])
plt.show()
```

```python
plt.plot(experiencia, salario_data_scientis, color = 'r', label = 'Data Scientist')
plt.plot(experiencia, salario_ing_software, color = 'g', label = 'Ing Software')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.legend()
plt.show()
```

```python
plt.plot(experiencia, salario_data_scientis, color = 'r', label = 'Data Scientist', linestyle = '--')
plt.plot(experiencia, salario_ing_software, color = 'g', label = 'Ing Software', linestyle = '--')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.legend()
plt.grid(color = 'b', linestyle = '--', linewidth = 2)
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [2, 6, 3, 5]
plt.bar(x, y)
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.show()
```

```python
x = ['A', 'B', 'C', 'D']
y = [2, 6, 3, 5]
tendencias = [5,3,7,8]
plt.bar(x, y)
plt.plot(x, tendencias, color = 'r')
plt.title('Salarios')
plt.xlabel('Experiencia')
plt.ylabel('Ganancias')
plt.show()
```

```python
# Gráfico de dispersión
import numpy as np
```

```python
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
plt.scatter(x, y, color = 'r', alpha = 0.5)
plt.title('Gráfico de dispersión')
plt.show
```

```python
datos = np.random.randn(10000)

plt.hist(datos, bins=10)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

```python
datos = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(datos, vert=True)
plt.show()
```

```python
import pandas as pd

data_range = pd.date_range(start='2023-01-01', end = '2023-12-31', freq='D')
precios = np.cumsum(np.random.randn(len(data_range))) + 100
```

```python
data = pd.DataFrame({'Date':data_range, 'Precios':precios})
data.head()
```

```python
plt.plot(data['Date'], data['Precios'], label = 'Precio')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
x1 = np.linspace(0, 10, 200)
y1 = np.sin(x)

categoría = ['a', 'b', 'c', 'd']
valores = [23, 65, 23, 75]

datos = [np.random.normal(0,std,1000) for std in range(1,3)]

x2 = np.random.rand(200)
y2 = np.random.rand(200)
```

```python
fig, axs = plt.subplots(2, 2, figsize = (15,10))

axs[0,0].plot(x, y1, label = 'seno')
axs[0,0].plot(x1, y2, label = 'Coseno')
axs[0,0].set_title('Gráfico de líneas')
axs[0,0].legend(loc = 'upper right')
axs[0,0].grid()

axs[0,1].bar(categoría, valores)
axs[0,1].set_title('Gráfico de barras')
axs[0,1].grid()

axs[1,0].boxplot(datos, vert = True)
axs[1,0].set_title('Gráfica de cajas y bigotes')
axs[1,0].grid()

axs[1,1].scatter(x2, y2)
axs[1,1].set_title('Gráfica de dispersión')
axs[1,1].grid()

fig.suptitle('Ejemplos de diferentes tipos de gráficos', fontsize = 22)
plt.show()
```
