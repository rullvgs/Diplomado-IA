## Algoritmos de ordenación

### Ordenación por burbuja


El método de ordenación por burbuja es uno de los algoritmos más conocidos y populares debido a su facilidad de comprensión y programación. Sin embargo, es uno de los menos eficientes, y normalmente se aprende su técnica, pero no se suele usar en la práctica debido a su baja eficiencia.

#### Descripción del Algoritmo

Dada una lista con `n` elementos, la ordenación por burbuja requiere hasta `n-1` pasadas. Por cada pasada se comparan elementos adyacentes y se intercambian sus valores si el primer elemento es mayor que el segundo. Al final de cada pasada, el elemento mayor "burbujea" hasta el final de la sublista actual.

#### Pasos del Algoritmo

1. **Pasada 0**: Se comparan elementos adyacentes `(A[0], A[1]), (A[1], A[2]), (A[2], A[3]), ..., (A[n-2], A[n-1])`. Se realizan `n-1` comparaciones, y por cada pareja `(A[i], A[i+1])`, se intercambian los valores si `A[i+1] < A[i]`. Al final de esta pasada, el elemento mayor de la lista está situado en `A[n-1]`.

2. **Pasada 1**: Se realizan las mismas comparaciones e intercambios, terminando con el elemento de segundo mayor valor en `A[n-2]`.

3. **Pasada 2**: Se realizan comparaciones hasta `A[n-3]`.

4. **Pasada n-1**: En esta pasada, el elemento más pequeño se almacena en `A[0]`.

#### Ejemplo

Vamos a ordenar el array `A = [50, 20, 40, 80, 30]` utilizando el algoritmo de burbuja. Introduciremos una variable `interruptor` para detectar si se ha producido un intercambio en la pasada.

   - Intercambio: `20` y `50`

   - Intercambio: `40` y `50`
   - Array: `[20, 40, 50, 80, 30]`
3. Comparamos `50` y `80`:
   - No hay intercambio
   - Array: `[20, 40, 50, 80, 30]`
4. Comparamos `80` y `30`:
   - Intercambio: `30` y `80`
   - Array: `[20, 40, 50, 30, 80]`
5. Al final de la pasada 0, el elemento mayor `80` está en la posición correcta.
6. `Interruptor = TRUE`
